# 3 Load some dummy data to the demo_user database

```python
host = 'host.docker.internal'
user = 'demo_user'
```

```python
import getpass
db_pw = getpass.getpass("Enter DB Password here")
```

```python
import tdmlplus #import needed to use widgets in teradataml
import teradataml as tdml
import os
import pandas as pd
```

```python
tdml.create_context(host, user, db_pw)
```

```python
all_csvs = os.listdir("sample_data_jaffleshop/")
for f in all_csvs:
    if f.endswith(".csv"):
        table_name = f.split(".csv")[0]
        df = pd.read_csv("sample_data_jaffleshop/"+f)
        # tdml.copy_to_sql(df, table_name, schema_name="aiworkbench_data", if_exists="replace")
        tdml.copy_to_sql(df, table_name, if_exists="replace")
        print(table_name, " uploaded")
```

```python
sql = """
select 
    t.databasename
    , t.tablename
from dbc.tablesv t
where databasename = \'demo_user\';
"""
```

```python
con = tdml.get_context()
table_list_df = pd.read_sql(sql, con)
```

```python
table_list_df
```

```python
# Display the uploaded tables in tabs
# limited to newer RDBMS versions
tdml.widgets.tab_dfs(table_names = [f.split(".csv")[0] for f in all_csvs] )
```
